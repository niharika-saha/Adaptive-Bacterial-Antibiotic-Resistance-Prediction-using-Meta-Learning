```python
#!/usr/bin/env python3
"""
experiment2_heldout_full.py

Simulated "Novel Mechanism / Drug Class Emergence" using a MEGARes-style CSV.

Models implemented:
- Prototypical Network (ProtoNet)
- RidgeMeta (MetaOptNet-like closed-form ridge solver)
- BayesianMeta (posterior predictive ridge)
- SupervisedFineTune (pretrain on non-heldout, finetune on held-out)

Outputs:
accuracy_vs_K.png, adaptation_supervised.png, tsne_protonet_task.png, results.json
"""
import os, random, math, json
from collections import Counter
from tqdm import trange, tqdm
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.nn.functional as F
from sklearn.manifold import TSNE
from sklearn.preprocessing import LabelEncoder

# -------------------------
# Config
# -------------------------
CSV_PATH = "megares_fasta_processed.csv"
HOLDOUT_FIELD = "drug_class"
HOLDOUT_NAMES = ["betalactams", "aminoglycosides", "glycopeptides"]

RANDOM_SEED = 42
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
OUTDIR = "exp2_outputs"
os.makedirs(OUTDIR, exist_ok=True)

# K-mer & model params
KMER_K = 5
MAX_KMER_VOCAB = 20000
ENC_HID = 512
ENC_OUT_DIM = 128
META_ITERS = 4000
PROTO_LR = 1e-3
RIDGE_LR = 1e-3
PRETRAIN_EPOCHS = 30
BATCH_SIZE_PRETRAIN = 64
N_WAY = 3
Q_QUERY = 5
KS = [1, 3, 5, 10]
TEST_TASKS_PER_K = 400
FINETUNE_STEPS = [0, 1, 5, 10, 50]
FINETUNE_LR = 1e-3

torch.manual_seed(RANDOM_SEED)
np.random.seed(RANDOM_SEED)
random.seed(RANDOM_SEED)

# -------------------------
# Utility functions
# -------------------------
def load_csv(path):
    df = pd.read_csv(path)
    req = {"gene_id","header","sequence","seq_len","gc_content","type","drug_class","mechanism","group"}
    if not req.issubset(df.columns):
        raise ValueError(f"CSV missing required columns, found {df.columns.tolist()}")
    return df

def kmer_counts(seq, k=KMER_K):
    seq = str(seq).upper().replace("U", "T")
    c = Counter()
    for i in range(max(0, len(seq)-k+1)):
        kmer = seq[i:i+k]
        if "N" not in kmer:
            c[kmer]+=1
    return c

def build_kmer_vocab(seqs, k=KMER_K, max_kmers=MAX_KMER_VOCAB):
    counts = Counter()
    for s in seqs:
        counts.update(kmer_counts(s, k))
    most = [k for k,_ in counts.most_common(max_kmers)]
    return {k:i for i,k in enumerate(most)}

def vectorize_kmer(seq, kmer_idx, k=KMER_K):
    v = np.zeros(len(kmer_idx), np.float32)
    for kmer,cnt in kmer_counts(seq,k).items():
        i=kmer_idx.get(kmer)
        if i is not None: v[i]=cnt
    s=v.sum()
    return v/s if s>0 else v

# -------------------------
# Dataset class
# -------------------------
class GroupDatabase:
    def __init__(self, df, groups, kmer_idx):
        self.kmer_idx=kmer_idx
        self.groups=[]
        self.group_to_seqs={}
        for g in groups:
            seqs=df.loc[df['group']==g,'sequence'].tolist()
            if len(seqs)>0:
                self.groups.append(g)
                self.group_to_seqs[g]=seqs

    def sample_task(self, N_way=N_WAY, K_shot=3, Q_query=Q_QUERY):
        classes=random.sample(self.groups, N_way)
        s_x,s_y,q_x,q_y=[],[],[],[]
        for i,g in enumerate(classes):
            seqs=self.group_to_seqs[g]
            total=K_shot+Q_query
            chosen=random.choices(seqs,k=total) if len(seqs)<total else random.sample(seqs,total)
            s,q=chosen[:K_shot],chosen[K_shot:]
            s_x+=[vectorize_kmer(x,self.kmer_idx) for x in s]
            q_x+=[vectorize_kmer(x,self.kmer_idx) for x in q]
            s_y+=[i]*K_shot
            q_y+=[i]*Q_query
        return (torch.tensor(np.stack(s_x)).float(),
                torch.tensor(s_y).long(),
                torch.tensor(np.stack(q_x)).float(),
                torch.tensor(q_y).long())

# -------------------------
# Models
# -------------------------
class SmallEncoder(nn.Module):
    def __init__(self, input_dim):
        super().__init__()
        self.net=nn.Sequential(nn.Linear(input_dim,ENC_HID),
                               nn.ReLU(),
                               nn.Linear(ENC_HID,ENC_OUT_DIM))
    def forward(self,x): return self.net(x)

class SupervisedModel(nn.Module):
    def __init__(self,enc,num_classes):
        super().__init__()
        self.encoder=enc
        self.fc=nn.Linear(ENC_OUT_DIM,num_classes)
    def forward(self,x): return self.fc(self.encoder(x))

# -------------------------
# Prototypical Network
# -------------------------
def proto_loss_and_acc(s_f,s_y,q_f,q_y,N_way=N_WAY):
    prototypes=[]
    for c in range(N_way):
        prototypes.append(s_f[s_y==c].mean(0))
    prototypes=torch.stack(prototypes)
    dists=torch.cdist(q_f,prototypes)
    logp=F.log_softmax(-dists,1)
    loss=F.nll_loss(logp,q_y)
    acc=(logp.argmax(1)==q_y).float().mean().item()
    return loss,acc

# -------------------------
# Ridge and Bayesian
# -------------------------
def ridge_closed_form_classifier(s_f,s_y,N_way,lambda_reg=1.0):
    D=s_f.size(1)
    Y=torch.zeros(len(s_y),N_way,device=s_f.device)
    for i in range(len(s_y)): Y[i,s_y[i]]=1.0
    X=s_f
    A=X.t()@X + lambda_reg*torch.eye(D,device=s_f.device) + 1e-6*torch.eye(D,device=s_f.device)
    rhs=X.t()@Y
    return torch.linalg.solve(A,rhs)

def ridge_predict(W,q_f): return q_f@W

def bayesian_posterior(s_f,s_y,N_way,prior_var=1.0,noise_var=1.0):
    D=s_f.size(1)
    Y=torch.zeros(len(s_y),N_way,device=s_f.device)
    for i in range(len(s_y)): Y[i,s_y[i]]=1.0
    X=s_f
    Sigma_inv=X.t()@X/noise_var + torch.eye(D,device=s_f.device)/prior_var
    rhs=X.t()@Y/noise_var
    W_mean=torch.linalg.solve(Sigma_inv,rhs)
    Sigma=torch.linalg.inv(Sigma_inv)
    return W_mean,Sigma

def bayesian_predict(W_mean,Sigma,q_f):
    logits=q_f@W_mean
    q_sigma=(q_f@Sigma*q_f).sum(1,keepdim=True)
    return logits/(1+torch.sqrt(q_sigma+1e-8))

# -------------------------
# Training loops
# -------------------------
def meta_train_protonet(encoder,db,iters=META_ITERS,lr=PROTO_LR):
    encoder.to(DEVICE)
    opt=torch.optim.Adam(encoder.parameters(),lr=lr)
    print("Meta-training ProtoNet...")
    for it in trange(iters):
        s_x,s_y,q_x,q_y=db.sample_task()
        s_x,q_x=s_x.to(DEVICE),q_x.to(DEVICE)
        s_y,q_y=s_y.to(DEVICE),q_y.to(DEVICE)
        enc_s,enc_q=encoder(s_x),encoder(q_x)
        loss,acc=proto_loss_and_acc(enc_s,s_y,enc_q,q_y)
        opt.zero_grad(); loss.backward(); opt.step()
        if it%200==0: print(f"it{it} loss={loss.item():.4f} acc={acc:.3f}")
    return encoder

def meta_train_ridgemeta(encoder,db,iters=META_ITERS,lr=RIDGE_LR):
    encoder.to(DEVICE)
    opt=torch.optim.Adam(encoder.parameters(),lr=lr)
    print("Meta-training RidgeMeta...")
    for it in trange(iters):
        s_x,s_y,q_x,q_y=db.sample_task()
        s_x,q_x=s_x.to(DEVICE),q_x.to(DEVICE)
        s_y,q_y=s_y.to(DEVICE),q_y.to(DEVICE)
        enc_s,enc_q=encoder(s_x),encoder(q_x)
        W=ridge_closed_form_classifier(enc_s,s_y,N_WAY)
        logits=ridge_predict(W,enc_q)
        loss=F.cross_entropy(logits,q_y)
        acc=(logits.argmax(1)==q_y).float().mean().item()
        opt.zero_grad(); loss.backward(); opt.step()
        if it%200==0: print(f"it{it} loss={loss.item():.4f} acc={acc:.3f}")
    return encoder

# -------------------------
# Supervised pretraining
# -------------------------
def pretrain_supervised(encoder,df_nonheld,kmer_idx,epochs=PRETRAIN_EPOCHS):
    groups=sorted(df_nonheld['group'].unique())
    le=LabelEncoder().fit(groups)
    X=np.stack([vectorize_kmer(s,kmer_idx) for s in df_nonheld['sequence']])
    y=le.transform(df_nonheld['group'])
    ds=torch.utils.data.TensorDataset(torch.tensor(X).float(),torch.tensor(y).long())
    dl=torch.utils.data.DataLoader(ds,batch_size=BATCH_SIZE_PRETRAIN,shuffle=True)
    model=SupervisedModel(encoder,len(groups)).to(DEVICE)
    opt=torch.optim.Adam(model.parameters(),lr=1e-3)
    loss_fn=nn.CrossEntropyLoss()
    for ep in range(epochs):
        tot_loss=0
        for bx,by in dl:
            bx,by=bx.to(DEVICE),by.to(DEVICE)
            loss=loss_fn(model(bx),by)
            opt.zero_grad(); loss.backward(); opt.step()
            tot_loss+=loss.item()
        if ep%5==0: print(f"epoch{ep} loss={tot_loss/len(dl):.4f}")
    return model,le

# -------------------------
# Evaluation helpers
# -------------------------
@torch.no_grad()
def evaluate_proto(encoder,db,K,ntasks=200):
    encoder.eval(); accs=[]
    for _ in range(ntasks):
        s_x,s_y,q_x,q_y=db.sample_task(K_shot=K)
        s_x,q_x=s_x.to(DEVICE),q_x.to(DEVICE)
        s_y,q_y=s_y.to(DEVICE),q_y.to(DEVICE)
        loss,acc=proto_loss_and_acc(encoder(s_x),s_y,encoder(q_x),q_y)
        accs.append(acc)
    return np.mean(accs),np.std(accs)/math.sqrt(len(accs))

@torch.no_grad()
def evaluate_ridge(encoder,db,K,ntasks=200):
    encoder.eval(); accs=[]
    for _ in range(ntasks):
        s_x,s_y,q_x,q_y=db.sample_task(K_shot=K)
        s_x,q_x=s_x.to(DEVICE),q_x.to(DEVICE)
        s_y,q_y=s_y.to(DEVICE),q_y.to(DEVICE)
        enc_s,enc_q=encoder(s_x),encoder(q_x)
        W=ridge_closed_form_classifier(enc_s,s_y,N_WAY)
        preds=ridge_predict(W,enc_q).argmax(1)
        acc=(preds==q_y).float().mean().item()
        accs.append(acc)
    return np.mean(accs),np.std(accs)/math.sqrt(len(accs))

@torch.no_grad()
def evaluate_bayesian(encoder,db,K,ntasks=200):
    encoder.eval(); accs=[]
    for _ in range(ntasks):
        s_x,s_y,q_x,q_y=db.sample_task(K_shot=K)
        s_x,q_x=s_x.to(DEVICE),q_x.to(DEVICE)
        s_y,q_y=s_y.to(DEVICE),q_y.to(DEVICE)
        enc_s,enc_q=encoder(s_x),encoder(q_x)
        W,S=bayesian_posterior(enc_s,s_y,N_WAY)
        preds=bayesian_predict(W,S,enc_q).argmax(1)
        acc=(preds==q_y).float().mean().item()
        accs.append(acc)
    return np.mean(accs),np.std(accs)/math.sqrt(len(accs))

# -------------------------
# Main
# -------------------------
def run_experiment():
    print("Loading CSV...")
    df=load_csv(CSV_PATH)
    held_mask=df[HOLDOUT_FIELD].isin(HOLDOUT_NAMES)
    held_groups=sorted(df[held_mask]['group'].unique())
    nonheld_df=df[~held_mask].copy()
    nonheld_groups=sorted(nonheld_df['group'].unique())
    print(f"Held-out {HOLDOUT_FIELD}s: {HOLDOUT_NAMES}")
    print(f"Held groups={len(held_groups)}, train groups={len(nonheld_groups)}")

    kmer_vocab=build_kmer_vocab(nonheld_df['sequence'],KMER_K)
    db_train=GroupDatabase(nonheld_df,nonheld_groups,kmer_vocab)
    db_test_held=GroupDatabase(df[held_mask],held_groups,kmer_vocab)
    sample_k=min(len(nonheld_groups),max(1,len(nonheld_groups)//5))
    normal_groups=random.sample(nonheld_groups,sample_k)
    db_test_normal=GroupDatabase(df[df['group'].isin(normal_groups)],normal_groups,kmer_vocab)

    results={}

    # ProtoNet
    proto_enc=SmallEncoder(len(kmer_vocab))
    proto_enc=meta_train_protonet(proto_enc,db_train)
    results["ProtoNet"]={}
    for K in KS:
        acc,se=evaluate_proto(proto_enc,db_test_held,K,TEST_TASKS_PER_K)
        results["ProtoNet"][K]=(acc,se)
        print(f"ProtoNet K={K} held={acc:.3f}")

    # RidgeMeta
    ridge_enc=SmallEncoder(len(kmer_vocab))
    ridge_enc=meta_train_ridgemeta(ridge_enc,db_train)
    results["RidgeMeta"]={}
    for K in KS:
        acc,se=evaluate_ridge(ridge_enc,db_test_held,K,TEST_TASKS_PER_K)
        results["RidgeMeta"][K]=(acc,se)
        print(f"RidgeMeta K={K} held={acc:.3f}")

    # Bayesian
    bayes_enc=SmallEncoder(len(kmer_vocab))
    bayes_enc=meta_train_ridgemeta(bayes_enc,db_train,iters=META_ITERS//2)
    results["BayesianMeta"]={}
    for K in KS:
        acc,se=evaluate_bayesian(bayes_enc,db_test_held,K,TEST_TASKS_PER_K)
        results["BayesianMeta"][K]=(acc,se)
        print(f"BayesianMeta K={K} held={acc:.3f}")

    # Supervised baseline
    sup_enc=SmallEncoder(len(kmer_vocab))
    sup_model,_=pretrain_supervised(sup_enc,nonheld_df,kmer_vocab)
    results["SupervisedFineTune"]={}
    for K in KS:
        accs=[]
        for _ in range(100):
            s_x,s_y,q_x,q_y=db_test_held.sample_task(K_shot=K)
            s_x,q_x=s_x.to(DEVICE),q_x.to(DEVICE)
            s_y,q_y=s_y.to(DEVICE),q_y.to(DEVICE)
            new_head=nn.Linear(ENC_OUT_DIM,N_WAY).to(DEVICE)
            opt=torch.optim.SGD(list(sup_model.encoder.parameters())+list(new_head.parameters()),lr=FINETUNE_LR)
            for _ in range(5):
                enc_s=sup_model.encoder(s_x)
                loss=F.cross_entropy(new_head(enc_s),s_y)
                opt.zero_grad(); loss.backward(); opt.step()
            preds=new_head(sup_model.encoder(q_x)).argmax(1)
            acc=(preds==q_y).float().mean().item()
            accs.append(acc)
        results["SupervisedFineTune"][K]=(np.mean(accs),np.std(accs)/math.sqrt(len(accs)))
        print(f"SupervisedFineTune K={K} held={np.mean(accs):.3f}")

    # Save & plot
    with open(os.path.join(OUTDIR,"results.json"),"w") as f: json.dump(results,f,indent=2)
    plt.figure(figsize=(6,4))
    for m in results:
        plt.errorbar(KS,[results[m][K][0] for K in KS],[results[m][K][1] for K in KS],label=m,marker='o')
    plt.xlabel("K (shots)")
    plt.ylabel("Held-out accuracy")
    plt.title(f"Held-out {HOLDOUT_FIELD}s: {', '.join(HOLDOUT_NAMES)}")
    plt.legend(); plt.grid(True); plt.tight_layout()
    plt.savefig(os.path.join(OUTDIR,"accuracy_vs_K.png"))
    print("Saved accuracy_vs_K.png")

if __name__=="__main__":
    run_experiment()
```
